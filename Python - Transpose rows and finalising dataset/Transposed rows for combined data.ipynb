```python
leak_types = ['Leak_Type1', 'Leak_Type2', 'Leak_Type3', 'Leak_Type4', 'Leak_Type5']
info = ['Organisation', 'Org_URL', 'Country', 'Industry', 'Pub_Date',
       'Leak_Type', 'Description', 'Ransom_URL', 'Ransom_Group', 'Views',
       'Leak_Size', 'Leak_Links', 'Org_Revenue', 'Percentage_Leaked']
# For cross referencing when generating rows (see the fourth row onward)
```

```python
df.iloc[0][info[2]] # extracting the country for row 0
```

```python
df_appended = [] # To store rows when leak type is split into rows
```

```python
for i in range(len(df)): # Loop the whole dataframe
    for leak in leak_types: # Loop the leak_types
        if (not(pd.isna(df.iloc[i][leak]))): # Check if the leak type is a null value and if it is, do not include it into the df_appended.
            row_details = { # Create a row
                "Organisation": df.iloc[i][info[0]],
                "Org_URL": df.iloc[i][info[1]],
                "Country": df.iloc[i][info[2]],
                "Industry": df.iloc[i][info[3]],
                "Pub_Date": df.iloc[i][info[4]],
                "Leak_Type": df.iloc[i][leak],
                "Description": df.iloc[i][info[6]],
                "Ransom_URL": df.iloc[i][info[7]],
                "Ransom_Group": df.iloc[i][info[8]],
                "Views": df.iloc[i][info[9]],
                "Leak_Size": df.iloc[i][info[10]],
                "Leak_Links": df.iloc[i][info[11]],
                "Org_Revenue": df.iloc[i][info[12]],
                "Percentage_Leaked": df.iloc[i][info[13]]
            }
            df_appended.append(row_details) # Insert new row into the table
```

```python
df_final = pd.DataFrame(df_appended) # Create dataframe for the appended rows
df_final = df_final.sort_values(by=['Ransom_Group', 'Pub_Date', 'Organisation'],ascending=True) # Sort the rows by ransom group, date published and organisation
df_final.fillna("NA", inplace=True) # Fill in empty information with NA (Not Applicable)
```
